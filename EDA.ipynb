1. Read the dataframe, check null value if present then do the needful, check duplicate row , if present then do the needful?
- DataFrame Processing Steps

Read the DataFrame
Check for Null Values
If null values are present, fill them appropriately.
Check for Duplicate Rows
If duplicates are found, remove them.
Implementation Code:

```python
import pandas as pd

# Step 1: Load the DataFrame from the CSV file
file_path = 'MultipleFiles/spotify.csv'
df = pd.read_csv(file_path)

# Step 2: Check for Null Values
if df.isnull().sum().any():
    print("Null values found. Filling null values...")
    # Example: Fill null values with the mean for numerical columns
    df.fillna(df.mean(), inplace=True)  # Adjust this method as needed
else:
    print("No null values found.")

# Step 3: Check for Duplicate Rows
if df.duplicated().sum() > 0:
    print("Duplicate rows found. Removing duplicates...")
    df.drop_duplicates(inplace=True)  # Remove duplicate rows
else:
    print("No duplicate rows found.")

# Display the cleaned DataFrame
print("Cleaned DataFrame:")
print(df.head())
```

Explanation:

Load the DataFrame: Reads the CSV file into a DataFrame.
Check for Null Values: Uses isnull().sum() to check for any null values. If found, it fills them with the mean of the respective columns.
Check for Duplicates: Uses duplicated().sum() to check for duplicate rows. If duplicates are found, it removes them using drop_duplicates().
Display Cleaned Data: Finally, it prints the first few rows of the cleaned DataFrame.

6. Among the top 5 most popular artists, which artist has the highest popularity on average? Calculate and display the average popularity for each artist?
-Among the top 5 most popular artists, which artist has the highest popularity on average? Calculate and
display the average popularity for each artist


```python
import pandas as pd

# Load the DataFrame from the CSV file
file_path = 'MultipleFiles/spotify.csv'
df = pd.read_csv(file_path)

# Step 1: Calculate average popularity for each artist
average_popularity = df.groupby('Artist')['Popularity'].mean().reset_index()

# Step 2: Sort the artists by average popularity and get the top 5
top_5_artists = average_popularity.sort_values(by='Popularity', ascending=False).head(5)

# Display the results
print("Top 5 Most Popular Artists and Their Average Popularity:")
print(top_5_artists)
```

Explanation:

Load Data: Reads the CSV file into a DataFrame.

Sort and Select Top 5: Sorts the artists by average popularity in descending order and selects the top 5.
Display Results: Prints the top 5 artists along with their average popularity.





7. For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist
- Most Popular Tracks for the Top 5 Most Popular Artists

To find the most popular track for each of the top 5 most popular artists, follow these steps:

Identify the Top 5 Most Popular Artists
Find the Most Popular Track for Each Artist
Display the Results

```python
import pandas as pd

# Load the DataFrame from the CSV file
file_path = 'MultipleFiles/spotify.csv'
df = pd.read_csv(file_path)

# Step 1: Calculate average popularity for each artist
average_popularity = df.groupby('Artist')['Popularity'].mean().reset_index()

# Step 2: Sort the artists by average popularity and get the top 5
top_5_artists = average_popularity.sort_values(by='Popularity', ascending=False).head(5)

# Step 3: Find the most popular track for each of the top 5 artists
most_popular_tracks = {}
for artist in top_5_artists['Artist']:
    most_popular_track = df[df['Artist'] == artist].sort_values(by='Popularity', ascending=False).iloc[0]
    most_popular_tracks[artist] = most_popular_track['Track Name']

# Display the results
print("Most Popular Tracks for the Top 5 Most Popular Artists:")
for artist, track in most_popular_tracks.items():
    print(f"{artist}: {track}")
```

Explanation:

Load Data: Reads the CSV file into a DataFrame.

9. Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or
violin plot?
-Exploring Variation of Track Duration Across Artists Using a Box Plot

To visualize how the duration of tracks varies across different artists, a box plot or violin plot can be used. Here’s how to do it with a box plot:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the DataFrame from the CSV file
file_path = 'MultipleFiles/spotify.csv'
df = pd.read_csv(file_path)

# Optional: Limit to top N artists by number of tracks for clarity in visualization
top_artists = df['Artist'].value_counts().nlargest(10).index
df_top_artists = df[df['Artist'].isin(top_artists)]

# Create the box plot for duration across artists
plt.figure(figsize=(12, 8))
sns.boxplot(data=df_top_artists, x='Artist', y='Duration (ms)', palette='coolwarm')
plt.title('Variation of Track Duration Across Top 10 Artists')
plt.xlabel('Artist')
plt.ylabel('Duration (ms)')
plt.xticks(rotation=45, ha='right')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Load Data: Reads the dataset into a DataFrame.
Filter Top Artists: Limits the plot to the top 10 artists with the most tracks for readability.
Create Box Plot:
Shows the distribution (median, quartiles, range, outliers) of track durations per artist.
Plot Styling: Includes rotated x-axis labels for better readability and gridlines for clarity.









10. How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a
violin plot.
-Visualizing the Distribution of Track Popularity for Different Artists Using a Violin Plot

To explore how the track popularity distribution varies among different artists, a violin plot is a great choice as it shows the distribution shape and density per artist.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the DataFrame from the CSV file
file_path = 'MultipleFiles/spotify.csv'
df = pd.read_csv(file_path)

# Optional: Focus on top 10 artists for clearer visualization
top_artists = df['Artist'].value_counts().nlargest(10).index
df_top_artists = df[df['Artist'].isin(top_artists)]

# Create a violin plot for popularity distribution by artist
plt.figure(figsize=(12, 8))
sns.violinplot(data=df_top_artists, x='Artist', y='Popularity', palette='Set2')
plt.title('Distribution of Track Popularity Among Top 10 Artists')
plt.xlabel('Artist')
plt.ylabel('Popularity')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Load Data: Read the CSV into a DataFrame.
Filter Top Artists: Use top 10 artists to keep the plot readable.
Violin Plot:
Displays the kernel density estimation of popularity per artist.
Reveals distribution shape, median, and variability.
Styling:
Rotated x-axis labels for readability.
Grid lines added for better visual guidance.
